# Analyzing Data

1- Analyze the overall loss/profit
2- Profit and loss by diff variables.
3- Areas where the bussiness is incurring losses or has low profits.


# Overall Profit/Loss

```python
profit_loss = round(df['Profit'].sum(), 2)
print('Overall profit/loss of the business is: $', profit_loss)
```

# Profit/Loss by Category

```python
#group the data by category and analyze the data by category

category_profit_loss= df.groupby('Category')['Profit'].sum().sort_values() 
plt.bar(category_profit_loss.index, category_profit_loss.values)
plt.title('Profit/Loss by Category')
plt.ylabel('Profit/Loss')
plt.xticks(rotation=90)
plt.show()
```

# Profit/Loss by Region

```python
# group the data by region and analyze the profit by region.

region_profit_loss = df.groupby('Region')['Profit'].sum().sort_values()

plt.figure(figsize=(12,6))
plt.bar(region_profit_loss.index, region_profit_loss.values)
plt.title("Profit/Loss by Region")
plt.xlabel("Region")
plt.ylabel("Profit/Loss")
plt.xticks(rotation=90)
plt.show()
```

# Profit/Loss by Sub-Category


```python
sub_category_profit_loss = df.groupby('Sub-Category')['Profit'].sum().sort_values()

plt.figure(figsize=(12,6))
plt.bar(sub_category_profit_loss.index, sub_category_profit_loss.values)
plt.title('Profit/Loss by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.xticks(rotation=90)
plt.show()
```

# Sales by Category

```python
import seaborn as sns
sns.barplot(x='Category', y='Sales', data=df)
```

# Sales by Region

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values()
plt.pie(region_sales.values, labels= region_sales.index, autopct='%1.1f%%' )
plt.title('Sales by Region')
plt.show()
```

# Profit/Loss by Discount

```python
discount_profit_loss = df.groupby('Discount')['Profit'].sum().sort_values()
discount_profit_loss.plot(kind='line')
plt.title('Profit and Loss by Discount')
plt.xlabel('Discount')
plt.ylabel('Profit/Loss')
plt.show()
```

# Profit & Sales by Ship Mode

```python
ship_mode_sales_profit = df.groupby('Ship Mode')[['Sales', 'Profit']].sum()
ship_mode_sales_profit.plot(kind='bar')
plt.title('Profit & Sales by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Amount')
plt.show()
```

# Sales & Profit by Segment

```python
segment_sales_profit = df.groupby('Segment')[['Sales', 'Profit']].sum()
segment_sales_profit.plot(kind='bar')
plt.title('Profit & Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Amount')
plt.show()
```

# Sales & Profit by State

```python
state_sales_profit = df.groupby('State')[['Sales', 'Profit']].sum()
state_sales_profit.plot(kind='bar', figsize=(16,4))
plt.title('Profit & Sales by State')
plt.xlabel('State')
plt.ylabel('Amount')
plt.show()
```
